### Pomodoro Technique

1. Identify a task or tasks to complete
2. Set timer for 25 minutes
3. Work on the task without distractions (phone, slack, etc)
4. When the timer goes off, take a 5 minute break
5. Repeat the process 3 more times
6. After four sessions, take a longer break... 15-30 minutes
7. Record each session with a tick or x in notebook

### Circadian Rhythm

1. Jot down your daily routine in hours, starting from waking up to going to bed
2. Use the following guide : 
  - On fire
  - Vibrant
  - Cruise control
  - 70%
  - distracted
  - slowing down
  - tired
  - hungry
3. Output
  - What is the most active part of your day 'on fire'?
  - what is the most 'distracted' part of your day
  - During which part of the day is it better to complete the hardest tasks
  - When is it best ot take a break
  - Who has similar working/relaxing rhythms in your team
